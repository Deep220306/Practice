```python
import threading
import time
```

```python
def print_numbers():
   for i in range(5):
    time.sleep(1)
    print(i)
def print_letters():
   for letter in ['a', 'b', 'c', 'd', 'e']:
    time.sleep(1)
    print(letter)
t = time.time()
print_numbers()
print_letters()
finished_time = time.time() - t
print(f"Finished in {finished_time} seconds") 
```

```python
def print_numbers():
   for i in range(5):
    time.sleep(1)
    print(i)
def print_letters():
   for letter in ['a', 'b', 'c', 'd', 'e']:
    time.sleep(1)
    print(letter)
#creating 2 threads
t1 = threading.Thread(target=print_numbers)
t2 = threading.Thread(target=print_letters)

t = time.time()
#starting threads
t1.start()
t2.start()

#waiting for threads to complete
t1.join()
t2.join()

finished_time = time.time() - t
print(f"Finished in {finished_time} seconds")
```

```python
import multiprocessing
import time

def square_numbers():
   for i in range(5):
    time.sleep(1)
    print(i*i)

def cube_numbers():
   for i in range(5):
    time.sleep(1.5)
    print(i*i*i)

if __name__ == "__main__":
    t = time.time()
    p1 = multiprocessing.Process(target=square_numbers)
    p2 = multiprocessing.Process(target=cube_numbers)
    
    p1.start()
    p2.start()
    
    p1.join()
    p2.join()
    
    finished_time = time.time() - t
    print(f"Finished in {finished_time} seconds")
```

```python
from concurrent.futures import ThreadPoolExecutor
import time

def print_no (no):
    time.sleep(1)
    return f"Number: {no}"
no = [1,2,3,4,5]

with ThreadPoolExecutor (max_workers = 3) as executor:
    results = executor.map(print_no, no)

for result in results:
    print(result)
```

```python
from concurrent.futures import ProcessPoolExecutor
import time

def sq_no (no):
    time.sleep(1)
    return f"Number: {no*no}"
no = [1,2,3,4,5]
if __name__ == "__main__":

    with ProcessPoolExecutor (max_workers = 3) as executor:
        results = executor.map(sq_no, no)

    for result in results:
        print(result)
```
