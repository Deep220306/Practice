```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("sales_data.csv")
df.head()
```

```python
df.describe()
```

```python
sales = df.groupby("Product_Category")["Sales_Amount"].sum()
sales
```

```python
plt.figure(figsize=(10, 6))
plt.bar(sales.index, sales.values, color='skyblue')
plt.title("Total Sales Amount by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Sales Amount")
plt.show()
```

```python
daily_sales = df.groupby("Sale_Date")["Sales_Amount"].sum()
daily_sales
```

```python
plt.figure(figsize=(10,5))
plt.plot(daily_sales.index, daily_sales.values, marker="o")
plt.title("Daily Sales")
plt.xlabel("Date")
plt.ylabel("Sales")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df["Unit_Price"], df["Sales_Amount"])
plt.xlabel("Unit Price")
plt.ylabel("Sales Amount")
plt.title("Sales Amount vs Unit Price")
plt.show()
```
