```python
import numpy as np
```

```python
arr1 = np.array([1, 2, 3, 4, 5])
print("Original array:", arr1)
print("Type of the array:", type(arr1))
print("Shape of the array:", arr1.shape)
```

```python
#reshape to 2d
arr2 = arr1.reshape(1, 5)
print("Reshaped array:", arr2)
```

```python
arr3 = np.array([[1, 2, 3], [4, 5, 6]])
print("Original 2D array:", arr3)
print("Shape of the 2D array:", arr3.shape)
```

```python
arr4=np.arange(0,10,2)
print("Array with arange:", arr4)
```

```python
#2d array of all 1
arr5=np.ones((2,3))
print("2D array of all ones:\n", arr5)
```

```python
#identity matrix
arr6=np.eye(3)
print("Identity matrix:\n", arr6)
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print("Array:\n", arr)
print("Shape:", arr.shape) # Output: (2, 3)
print("Number of dimensions:", arr.ndim) # Output: 2
print("Size (number of elements):", arr.size) # Output: 6
print("Data type:", arr.dtype) # Output: int64 (may vary based on platform)
print("Item size (in bytes):", arr.itemsize) # Output: 8 (may vary based on platform)
```
